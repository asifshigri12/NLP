```python
# Import necessary libraries
import pandas as pd
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
df = pd.read_csv('/content/SPAM text message 20170820 - Data.csv')
df.head()
```

##Data Preprocessing
Normalize the text data (lowercasing and punctuation removal)

```python
# Normalize text data
def normalize_text(text):
    text = text.lower()  # Lowercasing
    text = text.translate(str.maketrans('', '', string.punctuation))  # Removing punctuation
    return text

df['Message'] = df['Message'].apply(normalize_text)
```

```python
# Apply normalization to the message column
df['Message'] = df['Message'].apply(normalize_text)
df.head()
```

Tokenization and stopword removal
python


```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import nltk
```

```python
# Tokenization and stopword removal
stop_words = set(stopwords.words('english'))

def tokenize_and_remove_stopwords(text):
    tokens = word_tokenize(text)
    filtered_tokens = [word for word in tokens if word not in stop_words]
    return filtered_tokens

df['tokens'] = df['Message'].apply(tokenize_and_remove_stopwords)
```

Step3:  Transform the processed text data into numerical vectors using CountVectorizer
python

```python
df.columns
```

Step 4: Apply lemmatization

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
nltk.download('wordnet')
```

```python
# Initialize stemmer and lemmatizer
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
```

```python
# Apply stemming and lemmatization
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

def stem_and_lemmatize(tokens):
    stemmed = [stemmer.stem(word) for word in tokens]
    lemmatized = [lemmatizer.lemmatize(word) for word in stemmed]
    return lemmatized
```

```python
# Apply stemming and lemmatization
df['processed_tokens'] = df['tokens'].apply(stem_and_lemmatize)
df['processed_text'] = df['processed_tokens'].apply(lambda x: ' '.join(x))
```

```python
df.head()
```

```python
# Create N-gram models
vectorizer_unigram = CountVectorizer(ngram_range=(1, 1))
vectorizer_bigram = CountVectorizer(ngram_range=(1, 2))
vectorizer_trigram = CountVectorizer(ngram_range=(1, 3))
```

```python
# Transform text data
X_unigram = vectorizer_unigram.fit_transform(df['processed_text'])
X_bigram = vectorizer_bigram.fit_transform(df['processed_text'])
X_trigram = vectorizer_trigram.fit_transform(df['processed_text'])
```

```python
# Encode the labels
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(df['Category'])
```

Step 5: Create N-gram models (unigrams, bigrams, trigrams) and transform the text data

```python
# Create n-gram models
vectorizer_unigram = CountVectorizer(ngram_range=(1, 1))
vectorizer_bigram = CountVectorizer(ngram_range=(1, 2))
vectorizer_trigram = CountVectorizer(ngram_range=(1, 3))
```

```python
# Transform text data
X_unigram = vectorizer_unigram.fit_transform(df['processed_text'])
X_bigram = vectorizer_bigram.fit_transform(df['processed_text'])
X_trigram = vectorizer_trigram.fit_transform(df['processed_text'])
```

```python
df.head()
```

Step 6: Split the data into training and testing sets

```python
# Split the dataset into training and testing sets
X_train_uni, X_test_uni, y_train_uni, y_test_uni = train_test_split(X_unigram, y, test_size=0.2, random_state=42)
X_train_bi, X_test_bi, y_train_bi, y_test_bi = train_test_split(X_bigram, y, test_size=0.2, random_state=42)
X_train_tri, X_test_tri, y_train_tri, y_test_tri = train_test_split(X_trigram, y, test_size=0.2, random_state=42)
```

Step 7: Train a classifier on the training set
Using Multinomial Naive Bayes as the classifier

```python
# Initialize the model
model = MultinomialNB()
```

```python
# Train the model on unigrams
model.fit(X_train_uni, y_train_uni)
```

```python
# Train the model on bigrams
model_bi = MultinomialNB()
model_bi.fit(X_train_bi, y_train_bi)
```

```python
# Train the model on trigrams
model_tri = MultinomialNB()
model_tri.fit(X_train_tri, y_train_tri)
```

Step 8: Evaluate the model on the testing set

```python
# Evaluate the model on unigrams
y_pred_uni = model.predict(X_test_uni)
accuracy_uni = accuracy_score(y_test_uni, y_pred_uni)
precision_uni = precision_score(y_test_uni, y_pred_uni)
recall_uni = recall_score(y_test_uni, y_pred_uni)
f1_uni = f1_score(y_test_uni, y_pred_uni)
report_uni = classification_report(y_test_uni, y_pred_uni)
conf_matrix_uni = confusion_matrix(y_test_uni, y_pred_uni)

print("Unigram Model Performance:")
print(f"Accuracy: {accuracy_uni}")
print(f"Precision: {precision_uni}")
print(f"Recall: {recall_uni}")
print(f"F1-score: {f1_uni}")
print(f"Classification Report:\n {report_uni}")
print(f"Confusion Matrix:\n {conf_matrix_uni}")
```

```python
# Evaluate the model on bigrams
y_pred_bi = model_bi.predict(X_test_bi)
accuracy_bi = accuracy_score(y_test_bi, y_pred_bi)
precision_bi = precision_score(y_test_bi, y_pred_bi)
recall_bi = recall_score(y_test_bi, y_pred_bi)
f1_bi = f1_score(y_test_bi, y_pred_bi)
report_bi = classification_report(y_test_bi, y_pred_bi)
conf_matrix_bi = confusion_matrix(y_test_bi, y_pred_bi)

print("\nBigram Model Performance:")
print(f"Accuracy: {accuracy_bi}")
print(f"Precision: {precision_bi}")
print(f"Recall: {recall_bi}")
print(f"F1-score: {f1_bi}")
print(f"Classification Report:\n {report_bi}")
print(f"Confusion Matrix:\n {conf_matrix_bi}")
```

```python
# Evaluate the model on trigrams
y_pred_tri = model_tri.predict(X_test_tri)
accuracy_tri = accuracy_score(y_test_tri, y_pred_tri)
precision_tri = precision_score(y_test_tri, y_pred_tri)
recall_tri = recall_score(y_test_tri, y_pred_tri)
f1_tri = f1_score(y_test_tri, y_pred_tri)
report_tri = classification_report(y_test_tri, y_pred_tri)
conf_matrix_tri = confusion_matrix(y_test_tri, y_pred_tri)

print("\nTrigram Model Performance:")
print(f"Accuracy: {accuracy_tri}")
print(f"Precision: {precision_tri}")
print(f"Recall: {recall_tri}")
print(f"F1-score: {f1_tri}")
print(f"Classification Report:\n {report_tri}")
print(f"Confusion Matrix:\n {conf_matrix_tri}")
```
