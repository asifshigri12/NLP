```python
import pandas as pd
df = pd.read_csv('/content/Corona_NLP_train.csv', encoding = 'ISO-8859-1')
df.head()
```

```python
df_new = list(df['OriginalTweet'])
df_new
```

```python
len(df_new)
```

```python
from top2vec import Top2Vec
```

```python
model = Top2Vec(documents=df_new, speed='fast-learn', workers=4)
```

```python
model.get_num_topics()
```

```python
topic_sizes, topic_nums = model.get_topic_sizes()
```
